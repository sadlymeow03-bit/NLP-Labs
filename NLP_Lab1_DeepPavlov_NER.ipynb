```python
from deeppavlov import build_model, configs

# Загружаем конфигурацию и скачиваем веса модели
# Это может занять время (зависит от скорости интернета)
ner_model = build_model(configs.ner.ner_ontonotes_bert_mult, download=True)
```

## Демонстрация работы
Подаем на вход модели произвольный текст на русском языке. Ожидаем, что модель выделит человека (PERSON), локации (GPE) и организации (ORG).

```python
text = "Александр Пушкин родился в Москве, а учился в Царском Селе. Сейчас его стихи читают в компании Яндекс и во всем мире."

# Получаем предсказание
tokens, tags = ner_model([text])

print(f"{'Слово':<20} | {'Тег сущности'}")
print("-" * 35)

# Проходим циклом и фильтруем тег 'O'
for token, tag in zip(tokens[0], tags[0]):
    if tag != 'O':
        print(f"{token:<20} | {tag}")
```

```python
import pandas as pd
from datasets import load_dataset

print("Начинаю загрузку SberQuAD из Hugging Face...")

try:
    # Загружаем часть 'train' (обучающая выборка)
    # Выберем только первые 100 записей, чтобы не перегружать память
    dataset = load_dataset("sberquad", split='train[:100]')
    
    # Преобразуем структуру в плоскую таблицу Pandas
    # В SberQuAD ответы лежат в словаре {'text': [...], 'answer_start': [...]}, 
    # мы вытащим только сам текст ответа.
    df_sber = pd.DataFrame({
        'context': dataset['context'],
        'question': dataset['question'],
        'answer': [a['text'][0] for a in dataset['answers']],
        'source': 'SberQuAD'
    })

    print(f"Успешно загружено строк: {len(df_sber)}")
    print("\nПример данных (первые 3 строки):")
    display(df_sber.head(3))

except Exception as e:
    print(f"Произошла ошибка при загрузке: {e}")
    print("Если ошибка 'ModuleNotFoundError', значит нужно еще раз запустить установку pip install datasets")
```

```python
import pandas as pd
from datasets import load_dataset
import os

print("Начинаю загрузку ПОЛНОГО датасета SberQuAD")

try:
    # Загружаем весь обучающий набор (train)
    full_dataset = load_dataset("sberquad", split='train')
    
    # Преобразуем в DataFrame
    # Извлекаем текст ответа из вложенного словаря
    df_full_sber = pd.DataFrame({
        'context': full_dataset['context'],
        'question': full_dataset['question'],
        'answer': [a['text'][0] for a in full_dataset['answers']],
        'source': 'SberQuAD_Full'
    })

    print(f"Загрузка завершена! Всего строк: {len(df_full_sber)}")

    # СОХРАНЕНИЕ В ФАЙЛ
    file_path = 'sberquad_full_dataset.csv'
    df_full_sber.to_csv(file_path, index=False, encoding='utf-8')
    
    print(f"\nФайл '{file_path}' успешно создан в твоей папке!")
    print(f"Размер файла на диске: ~{os.path.getsize(file_path) / (1024*1024):.2f} MB")

    # Посмотрим на случайные 5 строк из середины
    display(df_full_sber.sample(5))

except Exception as e:
    print(f"Произошла ошибка: {e}")
```

```python
from datasets import load_dataset
import pandas as pd

print("Загружаю датасет MuSeRC (второй источник с вопросами и вариантами ответов)...")

try:
    # Загружаем MuSeRC из коллекции Russian SuperGLUE
    # Берем тренировочную часть
    muserc = load_dataset("russian_superglue", "muserc", split='train')

    muserc_data = []

    # Проходим по данным. В MuSeRC структура сложнее: 
    # Текст -> Список вопросов -> Список ответов (True/False)
    for example in muserc:
        context = example['paragraph']
        question = example['question']
        
        # Находим правильный ответ среди вариантов
        # В этом датасете 'answer' - это текст варианта, а 'label' (1 или 0) - верный он или нет
        correct_answer = ""
        all_options = []
        
        # В MuSeRC ответы привязаны к вопросам. Мы упростим структуру для датасета.
        # Для твоей курсовой это золото, т.к. тут есть и правильные, и неправильные варианты!
        answer_text = example['answer']
        is_correct = example['label']
        
        muserc_data.append({
            'context': context,
            'question': question,
            'answer': answer_text,
            'is_correct': is_correct, # 1 если верный, 0 если дистрактор (неверный)
            'source': 'MuSeRC'
        })

    # Создаем таблицу
    df_muserc = pd.DataFrame(muserc_data)
    
    # Чтобы не путаться, давай оставим только ПРАВИЛЬНЫЕ ответы (как в SberQuAD)
    # Но для курсовой ты потом сможешь вытащить и неверные!
    df_muserc_correct = df_muserc[df_muserc['is_correct'] == 1].head(10000)

    print(f"Успех! Собрано {len(df_muserc_correct)} пар вопрос-ответ из MuSeRC.")

    # СОХРАНЕНИЕ ВО ВТОРОЙ ФАЙЛ
    file_name = 'muserc_qa_dataset.csv'
    df_muserc_correct.to_csv(file_name, index=False, encoding='utf-8-sig')

    print(f"Файл '{file_name}' успешно создан.")
    display(df_muserc_correct.head(3))

except Exception as e:
    print(f"Ошибка при загрузке MuSeRC: {e}")
```

```python
import pandas as pd
from datasets import load_dataset
import os

print("Начинаю загрузку ПРОВЕРОЧНОГО датасета SberQuAD (Validation split)...")

try:
    # Загружаем проверочную часть (она содержит около 5 000 строк)
    val_dataset = load_dataset("sberquad", split='validation')
    
    # Преобразуем в таблицу
    df_val_sber = pd.DataFrame({
        'context': val_dataset['context'],
        'question': val_dataset['question'],
        'answer': [a['text'][0] for a in val_dataset['answers']],
        'source': 'SberQuAD_Validation'
    })

    print(f"Загрузка завершена! Всего строк: {len(df_val_sber)}")

    # СОХРАНЕНИЕ ВО ВТОРОЙ ФАЙЛ
    file_path_val = 'sberquad_val_dataset.csv'
    df_val_sber.to_csv(file_path_val, index=False, encoding='utf-8-sig')
    
    print(f"\nФайл '{file_path_val}' успешно создан в твоей папке!")
    print(f"Размер файла: ~{os.path.getsize(file_path_val) / (1024*1024):.2f} MB")

    # Посмотрим на первые 5 строк
    display(df_val_sber.head(5))

except Exception as e:
    print(f"Произошла ошибка: {e}")
```

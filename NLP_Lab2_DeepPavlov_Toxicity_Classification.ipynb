```python
import pandas as pd
import spacy
import sys
from deeppavlov import build_model

# 1. Установка и загрузка лемматизатора
# (выполняем установку прямо здесь на всякий случай)
!{sys.executable} -m spacy download en_core_web_sm

nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])

def lemmatize_en(text):
    if not isinstance(text, str): return ""
    doc = nlp(text.lower())
    return " ".join([token.lemma_ for token in doc])

# 2. Загрузка модели (используем имя, которое работало в твоем примере)
# Сначала установим её зависимости
!{sys.executable} -m deeppavlov install insults_kaggle_bert

# Теперь собираем модель
model_en = build_model("insults_kaggle_bert", download=True)

print("Модель загружена и лемматизатор готов!")
```

```python
import pandas as pd
from IPython.display import display

# 1. Загрузка датасета
# берем первые 500 строк, так как файл большой. Для демонстрации этого достаточно.
df = pd.read_csv('toxic_comments.csv', nrows=500, on_bad_lines='skip', engine='python')

# Определяем колонку с текстом
col = 'comment_text' if 'comment_text' in df.columns else 'text'
df = df.dropna(subset=[col])

# Для демонстрации в лабе возьмем 10 первых строк
df_demo = df.head(10).copy()

# 2. Применяем лемматизацию 
print("Запуск лемматизации...")
df_demo['lemmatized_text'] = df_demo[col].apply(lemmatize_en)

# 3. Классификация (Векторизация происходит внутри модели BERT)
print("Запуск модели классификации...")
# Модель вернет список меток, например ['Not Insult', 'Insult']
df_demo['prediction'] = model_en(df_demo['lemmatized_text'].tolist())

# 4. Красивый вывод результата
print("\nРезультаты Лабораторной работы №2:")
display(df_demo[[col, 'lemmatized_text', 'prediction']])
```
